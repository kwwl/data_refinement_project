```python
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/kemillamouri/Desktop/Data IA HETIC 2/data_refinement_project/DATA/RAW/dirty_cafe_sales.csv")
df_clean = df.copy()
df_clean.head()
```

```python
for col in df_clean.columns:
    df_clean[col] = df_clean[col].astype(str).str.strip()
```

```python
num_cols = ["Quantity", "Price Per Unit", "Total Spent"]

for col in num_cols:
    df_clean[col] = (
        df_clean[col]
        .str.replace(",", ".", regex=False)
    )
```

```python
for col in num_cols:
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")

df_clean[num_cols].head()
```

```python
df_clean["Transaction Date"] = pd.to_datetime(df_clean["Transaction Date"], errors="coerce")
df_clean["Transaction Date"].head()
```

```python
df_clean["Payment Method"] = df_clean["Payment Method"].replace({
    "credot card": "credit card",
    "creditcard": "credit card",
    "cash ": "cash",
    "cahs": "cash",
})

df_clean["Location"] = df_clean["Location"].replace({
    "dowtown": "downtown",
    "mid town": "midtown",
    "uptwon": "uptown",
})
```

```python
df_clean = df_clean.dropna(subset=["Transaction ID", "Quantity", "Price Per Unit"])
```

```python
df_clean = df_clean.dropna(subset=["Transaction Date"])
```

```python
df_clean = df_clean.drop_duplicates()
```

```python
df_clean = df_clean[(df_clean["Quantity"] > 0) & (df_clean["Quantity"] < 50)]
df_clean = df_clean[(df_clean["Price Per Unit"] > 0.5) & (df_clean["Price Per Unit"] < 50)]
```

```python
df_clean.info()
df_clean.describe(include="all")
```

```python
df_clean["Payment Method"] = df_clean["Payment Method"].replace("nan", "unknown")
df_clean["Location"] = df_clean["Location"].replace("nan", "unknown")
```

```python
df_clean["Payment Method"].value_counts()
df_clean["Location"].value_counts()
```

```python
df_clean["Location"] = df_clean["Location"].str.lower().str.strip()

df_clean["Location"] = df_clean["Location"].replace({
    "error": "unknown",
    "unknown": "unknown",
    "in-store": "in_store",
    "takeaway": "takeaway",
    "unknown": "unknown",
    "": "unknown"
})
```
