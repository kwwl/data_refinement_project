```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/Users/kemillamouri/Desktop/Data IA HETIC 2/data_refinement_project/DATA/PROCESSED/cafe_sales_clean.csv")
df.head()
```

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])
df["Quantity"] = df["Quantity"].astype(int)
df["Price Per Unit"] = df["Price Per Unit"].astype(float)
df["Total Spent"] = df["Total Spent"].astype(float)
df.info()
```

```python
df["Year"] = df["Transaction Date"].dt.year
df["Month"] = df["Transaction Date"].dt.month
df["Day"] = df["Transaction Date"].dt.day
df["Weekday"] = df["Transaction Date"].dt.day_name()
```

```python
# calcul panier moyen = Total Spent / Quantity par exemple
df["Unit Price Calculated"] = df["Total Spent"] / df["Quantity"]
```

```python
item_sales = df.groupby("Item").agg(
    total_quantity=pd.NamedAgg(column="Quantity", aggfunc="sum"),
    total_revenue=pd.NamedAgg(column="Total Spent", aggfunc="sum"),
    avg_price=pd.NamedAgg(column="Price Per Unit", aggfunc="mean")
).reset_index()

item_sales = item_sales.sort_values(by="total_revenue", ascending=False)
item_sales.head()
```

```python
# ventes mensuelles
monthly_sales = df.groupby("Month").agg(
    revenue=pd.NamedAgg(column="Total Spent", aggfunc="sum"),
    quantity=pd.NamedAgg(column="Quantity", aggfunc="sum")
).reset_index()

# ventes par jour de la semaine
weekday_sales = df.groupby("Weekday").agg(
    revenue=pd.NamedAgg(column="Total Spent", aggfunc="sum"),
    quantity=pd.NamedAgg(column="Quantity", aggfunc="sum")
).reindex(["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]).reset_index()

monthly_sales.head(), weekday_sales.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=item_sales.head(10), x="total_revenue", y="Item", palette="viridis")
plt.title("Top 10 des produits par chiffre d'affaires")
plt.xlabel("Chiffre d'affaires (â‚¬)")
plt.ylabel("Produit")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=monthly_sales, x="Month", y="revenue", palette="magma")
plt.title("Chiffre d'affaires par mois")
plt.xlabel("Mois")
plt.ylabel("CA (â‚¬)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=weekday_sales, x="Weekday", y="revenue", palette="coolwarm")
plt.title("Chiffre d'affaires par jour de la semaine")
plt.xlabel("Jour")
plt.ylabel("CA (â‚¬)")
plt.tight_layout()
plt.show()
```
